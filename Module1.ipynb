# **Q 1.5**

**Function to calculate the average price over the last N days**

```
#here N is the number of days
def avg_price(N):
    return (df['Average Price'].tail(N).sum())/N
```

**Average price over the course of last 1 week, 2 weeks, 1 month , 3 monts, 6 month and 1 year**

```
print("Average prices for :")
print("Last 1 week : ",avg_price(7))
print("Last 2 weeks : ",avg_price(14))
print("Last 1 month: ",avg_price(30))
print("Last 3 months : ",avg_price(90))
print("Last 6 months : ",avg_price(180))
print("Last 1 year : ",avg_price(365))
```

# **Q 1.7**

**Adding another column 'Trend' whose values depends on 'Day_Perc_Change'**

```
def trend(x): #function to fill up the value of coulm 'Trend'
  if x > -0.5 and x <= 0.5:
    return 'Slight or No change'
  elif x > 0.5 and x <= 1:
    return 'Slight Positive'
  elif x > -1 and x <= -0.5:
    return 'Slight Negative'
  elif x > 1 and x <= 3:
    return 'Positive'
  elif x > -3 and x <= -1:
    return 'Negative'
  elif x > 3 and x <= 7:
    return 'Among top gainers'
  elif x > -7 and x <= -3:
    return 'Among top losers'
  elif x > 7:
    return 'Bull run'
  elif x <= -7:
    return 'Bear drop'
df['Trend']= np.zeros(df['Day_Perc_Change'].count())
df['Trend']= df['Day_Perc_Change'].apply(lambda x:trend(x*100))
```

```
df.shape #(494, 20)Trend column is added
```
